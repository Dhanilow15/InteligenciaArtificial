### **Entrada 1:**
Georges Seurat – Bathers at Asnières, 1884 – óleo sobre tela - 201 cm × 300 cm – National Gallery, London.jpg

### **Saída:** 
Georges Seurat (Banhistas em Asniére)

### **Metodologia:** 
Seurat era um expoente do pontilhismo, técnica saída do movimento impressionista. Portanto, para identificar uma obra dele, a IA poderia analisar pixel a pixel a imagem, assim percebendo a ausência de linhas, uma vez que no pontilhismo a imagem é feita ponto a ponto, e suas cores são feitas justapondo uma paleta de cores primárias. Ou seja, uma pintura par categorizar uma pintura desse artista a IA deve procurar por uma obra pontilhada, colorida porém fria, com predominância de cores primárias e tons pastéis.

### **Entrada 2:**
Noite-Estrelada.-Vincent-van-Gogh.jpg

### **Saída:**
Vincent Van Gogh (Noite Estrelada)

### **Metodologia:** 
Em suas obras mais relevantes, Van Gogh sempre usou de longas e fortes pinceladas, mesclando grandes espirais, como na pintura acima, com linhas grossas. Além disso, a maioria de suas pinturas, se desenvolve a partir de uma paleta de cores derivada do azul, amarelo ou verde. Para caracterizar uma obra de Van Gogh, a IA deveria buscar por, longas retas e/ou espirais monocromáticas, que acabam por configurar contornos grossos nas figuras da composição. Além de uma paleta de cores não muito variada, fundamentada, na maioria das vezes, em uma das 3 cores supracitadas.

### **Entrada 3:**
degas_aula_de_danca_c.jpg

### **Saída:**
Edgar Degas (A aula de dança)

### **Metodologia:**
As obras de Degas são sempre caracterizadas pela representação de cenas do cotidiano, pintadas com grande complexidade e realismo. No que diz respeito as cores utilizadas, não há muita constância, já que no começo de sua carreira optou por usar mais tons neutros em suas composições, já no fim, partiu para uma abordagem oposta, utilizando cores vibrantes, com um grande contraste quase barroco.
Quase como um oposto a Van Gogh, para identificar uma obra desse artista a IA deveria buscar pela ausência de grandes linhas, uma vez que as obras de Degas são construídas com pinceladas mais suaves e pequenas, gerando contornos curvas e linhas menores que, na grande maioria dos casos, são quase imperceptíveis, afastando-se assim do pontilhismo. Talvez buscar por uma padronização no uso das cores gere resultados ambíguos, tendo em vista o vasto leque cromático abrangido pelo artista. Por fim, a presença de bailarinas com volumosos vestidos monocromáticos é um grande sinalizador de autoria.

Para detectar uma pintura de Berthe, a IA deve procurar por uma composição de cores predominantemente derivada de cores frias e tons pastéis, além disso a ausência de contornos é um grande indicador das obras dessa artista.

## **Metodologia:** 
  Com essa base de dados, a ideia seria classificar a raça do personagem no RPG Dungeons & Dragons 5ª Edição, de acordo com seus status iniciais.
<br>Sendo eles, os seguintes **dados de entrada**: altura, peso, velocidade, força, destreza, constituição, inteligência, sabedoria e carisma.
<br>Já as possíveis classificação de raça, ou seja, os **dados de saída**, do livro básico, são: anão, elfo, hafling (hobbit), humano, draconato (meio dragão), gnomo, meio-elfo, meio-orc e tiefling (meio demônio).
<br>
<br>
Nessa base, todos os status são numéricos. Altura, peso e velocidade são medidos em pés e libra (uma observação para a altura que está multiplicada por 10, para que todos os status sejam números inteiros). Já os demais, que são conhecidos como atributos, podem variar de 3 a 20 para um personagem de nível 1.
<br>
<br>
A previsão da saída vai se dar através da comparação desses status, uma vez que, cada raça tem características bem definidas e limitadas pelo livro. Por exemplo, os draconatos são a raça mais alta, a única com personagens de altura superior a 8 pés, já os gnomos e halflings são as raças mais baixas, tendo alturas em torno de 3 e 4 pés. <br>
Os atributos são gerados aleatoriamente, no caso desta base, os atributos foram obtidos randomicamente através da rolagem de dados, o método utilizado é o seguinte, são rolados 4 dados de 6 lados, o menor valor é descartado e os 3 restantes são somados e aplicados naquele status, resultando em valores entre 3 e 18. Entretanto, cada raça possui um bônus natural para um, ou mais status, o qual pode ser +1 ou +2 pontos, esse bônus é chamado de traço racial. Por exemplo, gnomos possuem +2 de inteligência, o que significa que um personagem gnomo nunca terá inteligência inferior a 5, e terá no máximo 20, já todos os seus demais status variam de 3 a 18, já os meio-orcs, possuem +2 de constituição e +1 de força, o que significa que um personagem meio-orc sempre terá seu atributo de constituição variando de 5 a 20, o de força de 4 a 19 e os demais atributos variando de 3 a 18. E assim vai seguindo, cada raça vai ter seu limite específico para cada atributo.
<br>
Com essas limitações em mente, a classificação seria feita.

```python
import pandas as pd
from sklearn import neighbors
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix

# carregando o dataset
dndStats = pd.read_csv('stats.csv')

print(dndStats)
```

```python
# Checando se há valores nulos ou duplicados que precisam ser tratados
print(f'Valores nulos:\n{dndStats.isnull().sum()}')
print(f'\nValores duplicados:\n{dndStats.duplicated().sum()}')
```

### **Diante disso não será necessário dropar nenhum objeto por valor ausente ou duplicado no pré-processamento!**

```python
# Checando correlação dos atributos
corr = dndStats.corr()
corr.style.background_gradient()
corr
```

### **Altura tem uma grande correlação com peso e velocidade, portanto, vamos dropar esse atributo para a análise**

```python
# dropando a altura
dndStats.drop(columns='height', inplace=True)
```

### **Checando os K-NN vizinhos**


```python
# separando o dataset para treinamento e teste
dndResults = dndStats['race']
dndSamples = dndStats.drop(columns='race')

xTrain, xTest, yTrain, yTest = train_test_split(dndStats, dndResults, test_size=0.1, random_state=69)

# separando os atributos dos resultados
xTrainResults = xTrain['race']
xTrainSamples = xTrain.drop(columns='race')

xTestResults = xTest['race']
xTestSamples = xTest.drop(columns='race')
```

### **Testando a matriz confusão com K = 1**

```python
# aplicando o método KNN
knn = neighbors.KNeighborsClassifier(n_neighbors=1)
knn.fit(xTrainSamples.values, xTrainResults.values)
testSample = (xTestSamples.values)
resultKNN = knn.predict(testSample)

# criando a matriz confusao
confusionMatrix = confusion_matrix(resultKNN, xTestResults.values)
print(confusionMatrix)
```

### **Testando a matriz confusão com K = 7**

```python
# aplicando o método KNN
knn = neighbors.KNeighborsClassifier(n_neighbors=7)
knn.fit(xTrainSamples.values, xTrainResults.values)
testSample = (xTestSamples.values)
resultKNN = knn.predict(testSample)

# criando a matriz confusao
confusionMatrix = confusion_matrix(resultKNN, xTestResults.values)
print(confusionMatrix)
```

### **Testando a matriz confusão com K = 7, e métrica de distância de manhattan**

```python
# aplicando o método KNN
knn = neighbors.KNeighborsClassifier(n_neighbors=7, p=1)
knn.fit(xTrainSamples.values, xTrainResults.values)
testSample = (xTestSamples.values)
resultKNN = knn.predict(testSample)

# criando a matriz confusao
confusionMatrix = confusion_matrix(resultKNN, xTestResults.values)
print(confusionMatrix)
```

### **Testando a matriz confusão com K = 7, e métrica de distância euclidiana**

```python
# aplicando o método KNN
knn = neighbors.KNeighborsClassifier(n_neighbors=7, p=2)
knn.fit(xTrainSamples.values, xTrainResults.values)
testSample = (xTestSamples.values)
resultKNN = knn.predict(testSample)

# criando a matriz confusao
confusionMatrix = confusion_matrix(resultKNN, xTestResults.values)
print(confusionMatrix)
```

### **Como amenizar o preconceito da IA?**
Primeiramente, devemos partir do fato que, a inteligência artificial em sí não é preconceituosa, ela trabalha em cima de dados coletados de um mundo preconceituoso. 
<br>
O preconceito pode estar imbutido não somente nesses dados, mas também na definição do problema que a IA deve resolver, problemas generalistas e nebulosos podem gerar resultados tendenciosos, independentemente da base que se é analisada. Portando, o primeiro passo para reduzir o preconceito da IA, é ter mais especificidade e transparência na hora de enquadrar os problemas que desejamos resolver, e também como desejamos resolvê-los. 
<br>Por exemplo, dada uma base de dados X, queremos escolher o novo garoto propaganda, para a propaganda de uma loja de pesca, na qual dono pediu que escolhessemos o mais bonito. Nesse caso, simplesmente ser bonito é uma característica muito generalista, podemos pensar que o dono da loja pensou nela pois talvez a beleza atraia mais clientes, mas a ideia genérica de belo, por si só, já traz um conceito preestabelecido, pois cada pessoa tem suas preferências. Provavelmente para escolher o modelo dessa loja, devemos escolher aquele que, tenha habilidades mais atrativas para o consumidor pesqueiro, como por exemplo, a pessoa que consegue por mais minhocas no anzol por minuto, ou o que pescou mais peixes no último ano, mesmo que ainda continue agregada a uma variável de beleza, essas especificidades já reduziriam possíveis problemas.
<br>
Por fim, o cuidado na hora da coleta dos dados para a análise também é de suma importância, tendo em vista que estes podem não representar a realidade, ou refletir preconceitos existentes. O seu preparo também é fundamental, uma vez que, considerar ou desconsiderar determinados atributos impacta fortemente no resultado final.
<br>
O preconceito da IA, antes de ser tecnológico, é um problema sobretudo humano, uma vez que a inteligência artificial, muitas vezes, carece de contexto social, e seus processos são imperfeitos. Devemos trabalhar para mitigar propensões explícitas e/ou ocultas nos dados de treinamento, pois no fim a justiça da IA é definida por nós mesmos.
